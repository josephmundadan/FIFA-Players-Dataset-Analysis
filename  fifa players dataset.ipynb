```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
fifa=pd.read_csv("C:\\Users\\josep\\Downloads\\fifa_data.csv")
```

```python
# 1.Which country has the most number of players (score :1)
```

```python
country_count=fifa['Nationality'].value_counts()
country_count
most_num_country=country_count.idxmax()
most_player_count=country_count.max()
print(f"the country with most number of player {most_num_country} with {most_player_count} players")
```

```python
# 2.Plot a bar chart of 5 top countries with the most number of players. (score :1)
```

```python
count_country=fifa["Nationality"].value_counts()
count=count_country.head(5)
plt.figure(figsize=(8,5),dpi=100)
plt.xlabel("country")
plt.ylabel("number of players")
plt.title("bar chart of players")
plt.bar(count.index,count.values,color="red")
plt.grid(axis="y",linestyle="--",alpha=0.7)
```

```python
3.Which player has the highest salary? (score :1)
```

```python
fifa= pd.read_csv("C:\\Users\\josep\\Downloads\\fifa_data.csv")

fifa['Wage'] = fifa['Wage'].str.replace('€', '').str.replace('K', '').astype(float)

highest_paid_player = fifa.loc[fifa['Wage'].idxmax(), ['Name', 'Wage']]

print(f"The player with the highest salary is {highest_paid_player['Name']}, with a wage of €{highest_paid_player['Wage']}K per week.")
```

```python
#4.Plot a histogram to get the salary range of the players.
```

```python
plt.figure(figsize=(8,5),dpi=300)
plt.xlabel("salary")
plt.ylabel("Number of Players")
plt.title("histogram chart")
plt.hist(fifa['Wage'], bins=30, color='skyblue', edgecolor='black')
plt.grid(axis='y',linestyle='--',alpha=0.7)          
```

```python
#Who is the tallest player in the fifa? (score :1)
```

```python
fifa['Height'] = fifa['Height'].astype(str)
fifa['Height'] = fifa['Height'].str.replace("'", ".").astype(float)
hight=fifa.loc[fifa["Height"].idxmax(),['Name','Height']]
print(f" the tallest player is {hight ['Name']} and height is {hight ['Height']} feet")
```

```python
club=fifa['Club'].value_counts()
most_num_player=club.idxmax()
print(f"the  club has the most number of players is { most_num_player}")
```

```python
#7.Which foot is most preferred by the players?Draw a bar chart for preferred foot (score :1)
```

```python
foot=fifa['Preferred Foot'].value_counts()
foot
max=foot.idxmax()
print(f"the more preferred by the players is {max}")
plt.figure(figsize=(8,5),dpi=500)
plt.xlabel("foot")
plt.ylabel("number")
plt.title("bar cahrt for Preferred Foot")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.bar(foot.index, foot.values, color="red")
```
